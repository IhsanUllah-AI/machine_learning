```python
import pandas as pd 
df=pd.DataFrame({
    'area':[2600,3000,3200,3600,4000],
    'price':[550000,565000,610000,680000,725000]
})
import matplotlib.pyplot as plt
plt.scatter(df['area'],df.price,color='green',marker='+')
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```

above + sign is the available data of dataframe or dataset

```python
import numpy as np
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(df[['area']],df['price'])
```

```python
x=[[3300,]]
pre=lr.predict(x)
print("predicted price for ",x," is =",pre)
```

```python
w=lr.coef_
b=lr.intercept_
f=w*3300+b
print(f)
```

```python
dff=pd.DataFrame({
    'area':[1000,1200,1500,1800,2200,2300,2800,3100,3400,3500,3900]
})
x=dff[['area']]
pre=lr.predict(x)
dff['predicted price']=pre
print(dff)
```

plot the predicted data along with available data

```python
plt.scatter(df['area'],df.price,color='green',marker='+')
plt.plot(dff.area,pre,color='blue')
plt.xlabel("area")
plt.ylabel("price")
plt.xlim(1000,4000)
plt.show()
```
