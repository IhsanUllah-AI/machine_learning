```python
import pandas as pd
import tensorflow as tf 
import keras as k 
from keras import layers
from keras import losses
from keras.activations import relu,sigmoid,linear,softmax 
from keras import Sequential
from sklearn.datasets import load_wine
```

```python
lw=load_wine()
dir(lw)
```

```python
df=pd.DataFrame(lw.data,columns=lw.feature_names)
df.head()
```

```python
df['target']=lw.target
df['target_name']=df.target.apply(lambda x:lw.target_names[x])
df.tail(2)
```

make independet and depenedet var

```python
x=df.drop(['target','target_name'],axis='columns')
y=df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20)
```

```python
model=Sequential(
    [
        layers.Dense(units=32,activation='relu',name='L1'),
        layers.Dense(units=16,activation='relu',name='L2'),
        layers.Dense(units=3,activation='softmax',name='L3')
    ]
)
```

```python
from keras import optimizers
```

```python
model.compile(loss=losses.SparseCategoricalCrossentropy(),metrics=['accuracy'],
              optimizer=optimizers.Adam(learning_rate=1e-3))
model.fit(x_train,y_train,epochs=9)
```

```python
model.predict(x_test)
```
